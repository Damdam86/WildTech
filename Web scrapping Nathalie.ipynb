```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from concurrent.futures import ThreadPoolExecutor, as_completed
import time
import pandas as pd
import json
from webdriver_manager.chrome import ChromeDriverManager

# Configuration du driver Chrome
chrome_options = Options()
chrome_options.add_argument("--headless")  # Mode headless (pas d'affichage du navigateur)

def setup_chrome_options():
    options = Options()
    options.add_argument("--headless")
    options.add_argument("--disable-gpu")
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")
    options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36")
    return options

def wait_for_element(driver, by, selector):
    try:
        element = WebDriverWait(driver).until(
            EC.presence_of_element_located((by, selector))
        )
        return element
    except:
        return None

def scrape_single_partner(link):
    # Cr√©er une nouvelle instance de webdriver pour chaque thread
    options = setup_chrome_options()
    service = Service(ChromeDriverManager().install())
    driver = webdriver.Chrome(service=service, options=options)

    try:
        print(f"üîó Acc√®s √† la page de la startup : {link}")
        driver.get(link)
        time.sleep(5)

        # Collecte des informations
        name_element = wait_for_element(driver, By.CSS_SELECTOR, "h1.topTitle")
        name = name_element.text.strip() if name_element else ""

        description_element = wait_for_element(driver, By.CSS_SELECTOR, "h3.topDescr")
        description = description_element.text.strip() if description_element else ""

        address_element = wait_for_element(driver, By.CSS_SELECTOR, "p.fulladdress")
        address = address_element.text.strip() if address_element else "Adresse non disponible"

        logo_element = wait_for_element(driver, By.CSS_SELECTOR, "div.startup-logo img")
        logo_url = logo_element.get_attribute("src") if logo_element else "Logo non disponible"

        website_element = wait_for_element(driver, By.CSS_SELECTOR, "div.right:nth-child(2) > p:nth-child(4) > a:nth-child(1)")
        website = website_element.get_attribute("href") if website_element else "Site non disponible"

        return {
            "name": name,
            "description": description,
            "address": address,
            "logo_url": logo_url,
            "website": website,
            "url": link
        }
    except Exception as e:
        print(f"Error processing {link}: {str(e)}")
        return None
    finally:
        driver.quit()

def scrape_partners_info_multithreaded(df_french_tech, max_threads=4):
    links = df_french_tech['link']
    french_data = []
    
    # Utiliser ThreadPoolExecutor pour ex√©cuter le scraping en parall√®le
    with ThreadPoolExecutor(max_threads) as executor:
        future_to_link = {executor.submit(scrape_single_partner, link): link for link in links}
        
        for future in as_completed(future_to_link):
            link = future_to_link[future]
            try:
                result = future.result()
                if result:
                    french_data.append(result)
                    print(f"Processed: {result['name']}")
                    
                    # Sauvegarde progressive
                    with open('french_tech.json', 'w', encoding='utf-8') as f:
                        json.dump(french_data, f, ensure_ascii=False, indent=4)
            except Exception as e:
                print(f"Error processing {link}: {str(e)}")
    
    return french_data

if __name__ == "__main__":
    # Scraping des pages principales pour obtenir les liens des startups
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)

    # Liste pour stocker les informations des startups
    startup_data = []

    # Nombre total de pages √† scraper
    TOTAL_PAGES = 1  # Change √† 175 pour tout scraper

    # Boucle pour scraper toutes les pages
    for page_num in range(1, TOTAL_PAGES + 1):
        print(f"üîç Scraping de la page {page_num}...")

        # URL de la page
        url = f"https://lespepitestech.com/best-of-pepites?created=3&page={page_num}"
        driver.get(url)

        try:
            # Attendre que la liste des startups soit charg√©e
            WebDriverWait(driver, 20).until(
                EC.presence_of_element_located((By.CSS_SELECTOR, "a.startup-entry-hitbox"))
            )
            print(f"‚úÖ Page {page_num} charg√©e avec succ√®s.")
        except Exception as e:
            print(f"‚ùå Erreur de chargement de la page {page_num}: {e}")
            continue  # Passer √† la page suivante

        # Extraire tous les liens des startups
        startup_links = [s.get_attribute("href") for s in driver.find_elements(By.CSS_SELECTOR, "a.startup-entry-hitbox")]

        # Cr√©er un DataFrame avec les liens des startups pour les passer √† la fonction multithreaded
        df_french_tech = pd.DataFrame({'link': startup_links})

        # Charger les donn√©es des startups en parall√®le avec le multithreading
        startup_data += scrape_partners_info_multithreaded(df_french_tech, max_threads=4)

    # üìå Sauvegarde des donn√©es
    df = pd.DataFrame(startup_data)
    df.to_csv("startup_info_pepites2.csv", index=False, encoding="utf-8")

    print(f"üéâ Donn√©es sauvegard√©es dans 'startup_info_pepites.csv'. Nombre de startups extraites : {len(startup_data)}")

    # Fermer le navigateur
    driver.quit()
```
